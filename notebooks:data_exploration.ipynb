```python

shipments["customs_clearance_time_days"] = (
    shipments["customs_clearance_time_days"]
    .fillna(shipments["customs_clearance_time_days"].median())
)
```

```python
shipments["customs_clearance_time_days"] = shipments["customs_clearance_time_days"].fillna(
    shipments["customs_clearance_time_days"].median()
)
```

```python
shipments["customs_clearance_time_days"] = pd.to_numeric(
    shipments["customs_clearance_time_days"],
    errors="coerce"
)
```

```python
median_clearance_time = shipments["customs_clearance_time_days"].median()

shipments["customs_clearance_time_days"] = (
    shipments["customs_clearance_time_days"]
    .fillna(median_clearance_time)
)
```

```python
print(shipments["customs_clearance_time_days"].dtype)
print(shipments["customs_clearance_time_days"].isnull().sum())
print("median used:", median_clearance_time) 
```
